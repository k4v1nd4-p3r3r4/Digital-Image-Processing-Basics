```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_rgb = cv2.imread('img_bgr.PNG')

# splitting image into different channels
r, g, b = cv2.split(img_rgb)

# Setting the grid size
plt.figure(figsize=(10, 10))

# plotting the original image
plt.subplot(221)
plt.title('Original')
plt.imshow(img_rgb)

# plotting the channels
plt.subplot(222)
plt.title('Blue Channel')
plt.imshow(b)

plt.subplot(223)
plt.title('Green Channel')
plt.imshow(g)

plt.subplot(224)
plt.title('Red Channel')
plt.imshow(r)
```

```python
img_hsv = cv2.imread('img_hsv.PNG');

#splitting imange to diffrent channel
h , s, v = cv2.split(img_hsv)
#setting the grid size
plt.figure(figsize=(10, 10))
#plotting the original image
plt.subplot(221)
plt.title('Original')
plt.imshow(img_hsv)
#plotting the channels
plt.subplot(222)
plt.title('Hue Channel')
plt.imshow(h)
plt.subplot(223)
plt.title('Saturation Channel')
plt.imshow(s)
plt.subplot(224)
plt.title('Value Channel')
plt.imshow(v)
```

```python
#splitting the image into diffrent channels
img = cv2.imread("img_hls.PNG");
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #convert to RGB

#convert RGB to HSL
img_hsl = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)

#splitting the image into diffrent channels
h, l, s = cv2.split(img_hsl)

#setting the grid size
plt.figure(figsize=(10, 10))
#plotting the original image
plt.subplot(221)
plt.title('Original')
plt.imshow(img_hsl)
#plotting the channels
plt.subplot(222)
plt.title('Hue Channel')
plt.imshow(h)
plt.subplot(223)
plt.title('Lightness Channel')
plt.imshow(l)
plt.subplot(224)
plt.title('Saturation Channel')
plt.imshow(s)

```
